```python
import random
def computer_guess():
    low= int(input("Enter the lowest number: "))
    high=int(input("Enter the highest number: "))
    guess = 0
    feedback=''
    while feedback!= 'c':
        if low != high:
            guess = random.randint(low, high)
        else:
            guess = low #can be also high
        feedback = input(f'Is {guess} too high(H) or too low(L) or Correct (C)?')
        if feedback == 'h':
            high =guess-1
        elif feedback == 'l':
            low =guess+1
    print(f'The computer guessed the number {guess} correctly.!')

computer_guess()
```
