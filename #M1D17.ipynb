```python
#Name_Cleaning
#df["Last_Name"] = df["Last_Name"].str.lstrip("...")
#df["Last_Name"] = df["Last_Name"].str.lstrip("/")
#df["Last_Name"] = df["Last_Name"].str.rstrip("_")

df["Last_Name"] = df["Last_Name"].str.strip("123._/")
df
```

```python
df["Phone_Number"]=df["Phone_Number"].str.replace(r"[-/|/]","",regex=True)

#df["Phone_Number"].apply(lambda x: x[0:3] + '_' + x[3:6]+'_'+ x[6:10])

#(df["Phone_Number"]).apply(lambda x:str(x))
df
```

```python
df["Phone_Number"] = df["Phone_Number"].astype(str)
```

```python
df["Phone_Number"] = (
    df["Phone_Number"].str[:3] + "-" +
    df["Phone_Number"].str[3:6] + "-" +
    df["Phone_Number"].str[6:]
)
```

```python
import numpy as np
df["Phone_Number"] = (
    df["Phone_Number"].replace(["nan--", "NaN","Na--"], np.nan)
    .dropna()
)
df
```

```python
df["Phone_Number"] = df["Phone_Number"].fillna("000-000-0000")
df
```

```python
df["Last_Name"] = df["Last_Name"].fillna("")
df
```

```python
import numpy as np
df["Address"] = (df["Address"].replace(["N/a"], np.nan)
    .dropna()
)
df
```

```python
df["Address"] = df["Address"].fillna("")
df
```

```python
df["Paying Customer"] = df["Paying Customer"].replace({"Y": "Yes", "N": "No", "N/a": ""})
df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].replace({"Y": "Yes", "N": "No", "NaN": ""})
df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].fillna("")
df
```

```python
df.drop("Not_Useful_Column", axis=1, inplace=True)
```

```python
df[["Street","State","Zip_code"]]=df["Address"].str.split(',',n=2,expand=True)
df
```

```python
df["State"] = df["State"].fillna("")
df["Zip_code"] = df["Zip_code"].fillna("")
df
```
