```python
subjects = ['Python', 'Phy', 'Chem', 'Math', 'Biology', 'CompSci', 'Geography', 'Sociology', 'Science', 'Arts', 'Commerce']
nums=[1,2,3,4,5]
```

```python
subjects.sort(reverse=True)
nums.sort(reverse=True)
print(subjects)
print(nums)
```

```python
sorted_subjects = sorted(subjects)
print(sorted_subjects)
```

```python
print(min(nums))
print(max(nums))
```

```python
nums.index(2)
```

```python
subjects.index('CompSci')
```

```python
for item in sorted_subjects:
    print(item)
```

```python
#list to string
sorted_subjects_str = ', '.join (sorted_subjects)
print(sorted_subjects_str)
```

```python
#string to list
new_list = sorted_subjects_str.split(',')
print(new_list)
```
