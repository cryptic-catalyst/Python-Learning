```python
#file opening
f = open('test.txt','r')
print(f.mode)
f.close()
```

```python
#contents manager style open and auto closing files
with open('test.txt','r') as f:
     f_contents = f.read()
     print(f_contents)

```

```python
#Write file-copy
with open('test.txt','r') as rf:
    with open('test_copy.txt','w') as wf:
        for line in rf:
            wf.write(line)
```

```python
#image copy for image ise rb & wb (b for Binary)
#Write file-copy
with open('girl.jpg','rb') as rf:
    with open('girl_copy.jpg','wb') as wf:
        for line in rf:
            wf.write(line)
```

```python
#Copy with Chunk
with open('girl.jpg','rb') as rf:
    with open('girl_copy2.jpg','wb') as wf:
        chunk_size = 1024
        rf_chunk = rf.read(chunk_size)
        while len(rf_chunk)>0:
            wf.write(rf_chunk)
            rf_chunk = rf.read(chunk_size)

```
