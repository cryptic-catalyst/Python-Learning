```python
words = [
"Horse","Dog","Cat","Elephant","Tiger","Lion","Monkey","Bear","Wolf","Fox",
"Cow","Goat","Sheep","Deer","Rabbit","Mouse","Rat","Squirrel","Kangaroo","Panda",
"Zebra","Giraffe","Leopard","Cheetah","Camel","Donkey","Pig","Chicken","Duck","Eagle",
"Falcon","Owl","Parrot","Swan","Peacock","Shark","Whale","Dolphin","Octopus","Crab",
"Lobster","Fish","Frog","Snake","Lizard","Turtle","Crocodile","Alligator","Ant","Bee",
"Butterfly","Spider","Beetle","Worm","Tree","Flower","Grass","Leaf","Root","Seed",
"Forest","Mountain","River","Lake","Ocean","Island","Desert","Valley","Hill","Field",
"Road","Bridge","Building","House","Apartment","School","College","University","Hospital","Market",
"Shop","Office","Factory","Farm","Garden","Park","Library","Museum","Temple","Church",
"Mosque","Stadium","Airport","Station","Hotel","Restaurant","Cafe","Bank","Court","Prison",
"Book","Pen","Pencil","Paper","Notebook","Table","Chair","Bed","Sofa","Door",
"Window","Wall","Floor","Roof","Clock","Phone","Computer","Laptop","Keyboard","Mouse",
"Screen","Camera","Television","Radio","Speaker","Bottle","Cup","Plate","Spoon","Fork",
"Knife","Bag","Box","Basket","Bucket","Car","Bus","Train","Bicycle","Motorcycle",
"Truck","Boat","Ship","Airplane","Helicopter","Rocket","Engine","Wheel","Ticket","Passport",
"Money","Coin","Card","Key","Lock","Ring","Watch","Glasses","Shoes","Shirt",
"Pants","Jacket","Hat","Scarf","Gloves","Food","Rice","Bread","Milk","Water",
"Tea","Coffee","Sugar","Salt","Fruit","Apple","Banana","Mango","Orange","Grapes",
"Vegetable","Potato","Tomato","Onion","Carrot","Meat","Fish","Egg","Cake","Chocolate",
"Music","Song","Movie","Game","Sport","Ball","Goal","Team","Player","Teacher",
"Student","Doctor","Nurse","Engineer","Artist","Writer","Farmer","Driver","Manager","Worker"
]

import random
import string

def get_valid_word(words):
    word = random.choice(words)
    while '_' in word or '-' in word or ' ' in word:
        word = random.choice(words)
    return word.lower()

def hangman():
    word = get_valid_word(words)
    word_letters = set(word)
    alphabet = set(string.ascii_lowercase)
    used_letters = set()

    lives = 6

    while len(word_letters) > 0 and lives>0 :
        print('You have ',{lives},' left, you used the following letters:', ' '.join(used_letters))

        word_list = [letter if letter in used_letters else '_' for letter in word]
        print('Your word is:', ' '.join(word_list))

        user_letter = input('Guess a letter: ').lower()

        if user_letter in alphabet - used_letters:
            used_letters.add(user_letter)
            if user_letter in word_letters:
                word_letters.remove(user_letter)
            else:
                lives = lives-1
                print('Letter is not in the Word')
        elif user_letter in used_letters:
            print('You have already used the letter, please try again.')
        else:
            print('Invalid input, please try again.')
    if lives == 0:
        print('NO LIVES LEFT, GAME OVER\nThe word was: ', word)
    else:
        print("Congratulations! You guessed the word:", word)
hangman()
```
