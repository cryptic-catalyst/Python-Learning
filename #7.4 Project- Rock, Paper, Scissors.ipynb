```python
import random

def play():
    user = input ("Input your choice: \n'r' for rock, 'p' for paper, 's' for scissors ")
    computer = random.choice(['r','p', 's'])

    if user == computer:
        return 'Match tie'
    # r>s, s>p, p>r
    if is_win(user, computer):
        return 'You won!'
    else:
        return 'You lost!'

def is_win(player, opponent):
    if (player == 'r' and opponent == 's') or (player == 's' and opponent == 'p') or (player == 'p' and opponent == 'r'):
        return True
    else:
        return False
print(play())

```
